```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Data (x = 2, y = 4)
x = torch.tensor([2.0])
y = torch.tensor([4.0])

# Define the model class
class SimpleLinearModel(nn.Module):
    def __init__(self):
        super(SimpleLinearModel, self).__init__()
        # Initialize the weight (w12)
        self.w12 = nn.Parameter(torch.tensor([0.5]))  # Initial weight

    def forward(self, x):
        # Define the forward pass: linear relationship
        return self.w12 * x  # No bias is used for simplicity

# Instantiate the model
model = SimpleLinearModel()

# Define the loss function (Mean Squared Error)
criterion = nn.MSELoss()

# Define the optimizer (Stochastic Gradient Descent) with learning rate alpha = 0.5
optimizer = optim.SGD(model.parameters(), lr=0.5)

# Training loop
epochs = 100
loss_values = []  # List to store the loss values over epochs

for epoch in range(epochs):
    # Forward pass: compute the predicted value
    y_pred = model(x)

    # Compute the loss
    loss = criterion(y_pred, y)

    # Backward pass: compute gradients
    optimizer.zero_grad()
    loss.backward()

    # Update the weights
    optimizer.step()

    # Store the loss value for plotting
    loss_values.append(loss.item())

    # Print the loss every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}, Weight: {model.w12.item():.4f}")

# Plot the loss over epochs
plt.plot(range(epochs), loss_values, label="Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss over Epochs")
plt.legend()
plt.show()

```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Data (x = 2, y = 4)
x = torch.tensor([2.0])
y = torch.tensor([4.0])


# Define the model class
class SimpleLinearModel(nn.Module):
    def __init__(self):
        super(SimpleLinearModel, self).__init__()
        # Initialize the weight (w12)
        self.w12 = nn.Parameter(torch.tensor([0.0]))  # Initial weight

    def forward(self, x):
        # Define the forward pass: linear relationship
        return self.w12 * x  # No bias is used for simplicity


# Instantiate the model
model = SimpleLinearModel()

# Define the loss function (Mean Squared Error)
criterion = nn.MSELoss()

# Define the optimizer (Stochastic Gradient Descent) with a smaller learning rate
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training loop
epochs = 30
loss_values = []  # List to store the loss values over epochs
weight_history = []  # Store weight history for visualization

for epoch in range(epochs):
    # Forward pass: compute the predicted value
    y_pred = model(x)

    # Compute the loss
    loss = criterion(y_pred, y)

    # Backward pass: compute gradients
    optimizer.zero_grad()
    loss.backward()

    # Update the weights
    optimizer.step()

    # Store the loss and weight value for plotting
    loss_values.append(loss.item())
    weight_history.append(model.w12.item())

    # Print epoch, loss, and the model output
    print(
        f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}, Predicted Output: {y_pred.item():.4f}"
    )

# Visualization of the Loss Landscape
# Create a range of weight values for plotting the loss curve
weight_range = np.linspace(-2, 6, 100)
loss_surface = []

for w in weight_range:
    model.w12.data = torch.tensor([w])  # Set the weight
    with torch.no_grad():
        y_pred_surface = model(x)
        loss_surface.append(criterion(y_pred_surface, y).item())

# Convert to numpy for plotting
loss_surface = np.array(loss_surface)

# Plot the loss over epochs
plt.figure(figsize=(12, 6))

# Loss Surface
plt.subplot(1, 2, 1)
plt.plot(weight_range, loss_surface, label="Loss Surface", color="blue")
plt.plot(
    weight_history,
    [criterion(torch.tensor([w]) * x, y).item() for w in weight_history],
    "ro-",
    label="Gradient Descent Path",
)
plt.title("Loss Surface with Gradient Descent Path")
plt.xlabel("Weight (w12)")
plt.ylabel("Loss")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(range(epochs), loss_values, label="Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss over Epochs in Gradient Descent")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Define the input and output
X = 2
y = 4

# Initialize the weight
w = 0.5

# Learning rate
alpha = 0.1

# Number of iterations
iterations = 5

# Training loop
for i in range(iterations):
    # Forward pass
    y_pred = w * X
    loss = 0.5 * (y - y_pred) ** 2
    print(f"Iteration {i + 1}:")
    print(f"Predicted y: {y_pred}")
    print(f"Loss: {loss}")

    # Backpropagation
    gradient = (y - y_pred) * X
    w = w + alpha * gradient
    print(f"Updated weight: {w}\n")

# Final weight
print(f"Final weight: {w}")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

x = torch.tensor([2.0])
y = torch.tensor([4.0])


class SimpleLinearModel(nn.Module):
    def __init__(self):
        super(SimpleLinearModel, self).__init__()
        self.w12 = nn.Parameter(torch.tensor([0.0]))

    def forward(self, x):
        return self.w12 * x


model = SimpleLinearModel()
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

epochs = 50
loss_values = []
weight_history = []
predicted_values = []

model.w12.data = torch.tensor([0.0])

for epoch in range(epochs):
    y_pred = model(x)
    loss = criterion(y_pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    loss_values.append(loss.item())
    weight_history.append(model.w12.item())
    predicted_values.append(y_pred.item())

    print(
        f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}, Predicted Output: {y_pred.item():.4f}"
    )

weight_range = np.linspace(-2, 6, 100)
loss_surface = []

for w in weight_range:
    model.w12.data = torch.tensor([w])
    with torch.no_grad():
        y_pred_surface = model(x)
        loss_surface.append(criterion(y_pred_surface, y).item())

loss_surface = np.array(loss_surface)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(weight_range, loss_surface, label="Loss Surface", color="blue")
plt.plot(
    weight_history,
    [criterion(torch.tensor([w]) * x, y).item() for w in weight_history],
    "ro-",
    label="Gradient Descent Path",
)
plt.title("Loss Surface with Gradient Descent Path")
plt.xlabel("Weight (w12)")
plt.ylabel("Loss")
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(range(epochs), loss_values, label="Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss over Epochs in Gradient Descent")
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(range(epochs), predicted_values, label="Predicted Output", color="orange")
plt.axhline(y=4.0, color="red", linestyle="--", label="Actual Output (y=4)")
plt.xlabel("Epochs")
plt.ylabel("Output")
plt.title("Predicted Output Approaching Actual Output")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
